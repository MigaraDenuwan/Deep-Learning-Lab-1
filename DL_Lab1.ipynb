```python
# Task 1.2: compare distributions with a histogram
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

exp = np.random.exponential(scale=1.0, size=100000)
uni = np.random.rand(100000)
norm = np.random.randn(100000)

plt.figure(figsize=(9,5))
plt.hist(exp, bins=100, density=True, histtype='step', label='exponential')
plt.hist(uni, bins=100, density=True, histtype='step', label='uniform')
plt.hist(norm, bins=100, density=True, histtype='step', label='normal')
plt.legend()
plt.title("Random distributions (100000 samples each)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.xlim(-5, 10)
plt.ylim(0, 1.2)
plt.show()
```

```python
# Task 1.3: 3D surface plot Z = X^2 + Y^2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')
fig.colorbar(surf, shrink=0.5)
plt.show()
```

```python
# Task 1.4: correlation heatmaps (pearson and spearman)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Option A: If you have your own CSV, upload it and load with pd.read_csv('filename.csv')
df = pd.read_csv('Seaborn Tutorial.ipynb')

# Option B (quick demo): use seaborn built-in dataset so you can run this right now
# df = sns.load_dataset('penguins').dropna()

# Pick numeric columns (replace these names with your actual columns if you have them)
num_cols = df.select_dtypes(include='number').columns.tolist()  # list of numeric columns
print("Numeric columns used:", num_cols)

data = df[num_cols]  # select the numeric columns

pearson = data.corr(method='pearson')
spearman = data.corr(method='spearman')

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.heatmap(pearson, annot=True, fmt=".2f")
plt.title("Pearson correlation")

plt.subplot(1,2,2)
sns.heatmap(spearman, annot=True, fmt=".2f")
plt.title("Spearman correlation")
plt.tight_layout()
plt.show()
```
